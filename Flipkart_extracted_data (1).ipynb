# Flipkart Laptop Data Extraction

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/laptops/pr?sid=6bo,b5g&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_2454b088-8975-47b8-a816-868f425e21cd_1_372UD5BXDFYS_MC.F6KJFVU7EOGS&otracker=hp_rich_navigation_1_1.navigationCard.RICH_NAVIGATION_Electronics~Laptop%2Band%2BDesktop~All_F6KJFVU7EOGS&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_1_L2_view-all&cid=F6KJFVU7EOGS&page=1"
```

```python
req = requests.get(url)
```

```python
content = BeautifulSoup(req.content, 'html.parser')
```

# Actual Code : 

```python
Laptop_features_list=[]
Lname=[]
Lprice=[]
Linformation =[]
lst2 = []


for page_no in range(1,45):
  url = 'https://www.flipkart.com/laptops/pr?sid=6bo,b5g&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_2454b088-8975-47b8-a816-868f425e21cd_1_372UD5BXDFYS_MC.F6KJFVU7EOGS&otracker=hp_rich_navigation_1_1.navigationCard.RICH_NAVIGATION_Electronics~Laptop%2Band%2BDesktop~All_F6KJFVU7EOGS&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_1_L2_view-all&cid=F6KJFVU7EOGS&page='+str(page_no)
  req = requests.get(url)
  content = BeautifulSoup(req.content, 'html.parser')

  MOBILE = content.find_all('div',class_='_2kHMtA')

  for i in MOBILE:
    Lname.append(i.find_all('div', class_='_4rR01T')[0].text)
    Lprice.append(i.find_all('div', class_='_30jeq3 _1_WHN1')[0].text)
    common_class_of_features = i.find_all('li', class_='rgWa7D')
    Laptop_features_list.append(common_class_of_features)

    
```

# Removing the HTML text from 'Laptop_features_list' : 

```python
Final_laptop_features_list = []
for i in range(len(Laptop_features_list)):
  lst1 = []
  for j in range(len(Laptop_features_list[i])):
    lst1.append((Laptop_features_list[i][j]).text)
  Final_laptop_features_list.append(lst1) 
```

```python
len(Final_laptop_features_list)
```

```python
Final_laptop_features_list[0:5]
```

# Creating DataFrame : 

```python
DATA = {'LAPTOP NAME': Lname, 'PRICE': Lprice, 'LAPTOP FEATURES': Final_laptop_features_list}
df = pd.DataFrame(DATA)
```

## processors : 

```python
processors = []
for i in df['LAPTOP FEATURES'].iteritems():
  x1=[]
  for j in range(len(i[1])):
    if "HDD" in i[1][j] or "SSD" in i[1][j]:
      x1.append(i[1][j])
  processors.append(x1)
#processors
len(processors)
```

```python
processors
```

```python
dict1 = {'PROCESSORS':processors}
df1 = pd.DataFrame(dict1)
```

```python
df1.head(2)
```

```python
df1['New_ID'] = df.index 
```

```python
df2 = pd.merge(df,df1,on='New_ID')
df2.head()
```
